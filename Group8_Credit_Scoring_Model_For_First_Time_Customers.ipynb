```python
#Importing the 6 month history of credit payments by each customer
import pandas as pd
import numpy as np
credit = pd.read_csv('./data/credit-record-final.csv')
credit
```

```python
#Understanding the Credit Record data
credit_record_dictionary = pd.read_csv('./data/credit-record-dictionary.csv')
credit_record_dictionary
```

```python
#Setting a clear evaluation criteria for assigning credit score points out of 100 maximum
evalution_criteria_dictionary = pd.read_csv('./data/evaluation-criteria-dictionary.csv')
evalution_criteria_dictionary
```

```python
#Assigning the credit score points based on the status for each month
```

```python
conditions=[
    (credit['STATUS']=='C'),
    (credit['STATUS']=='0'),
    (credit['STATUS']=='X'),
    (credit['STATUS']=='1'),
    (credit['STATUS']=='2'),
    (credit['STATUS']=='3'),
    (credit['STATUS']=='4'),
    (credit['STATUS']=='5')
]
```

```python
values=['60','60', '50','50','40','30','20','0']
```

```python
credit['CREDIT_SCORE'] = np.select(conditions,values)
```

```python
credit
```

```python
#Verify the data type for the new Credit Score value, then change to an integer
credit['CREDIT_SCORE'].dtype
```

```python
credit['CREDIT_SCORE']=credit['CREDIT_SCORE'].astype(int)
credit['CREDIT_SCORE'].dtype
```

```python
credit['ID']=credit['ID'].astype(int)
credit['ID'].dtype
```

```python
credit2=credit[['ID', 'CREDIT_SCORE']].copy()
credit2
```

```python
credit2['ID']=credit2['ID'].astype(int)
credit2['CREDIT_SCORE']=credit2['CREDIT_SCORE'].astype(int)
```

```python
#Calculate the average credit score based on the points assigned to each of the 6 months
credit3=credit2.groupby('ID')['CREDIT_SCORE'].mean().round(2)
credit3
```

```python
credit4=credit3.to_frame()
credit4
```

```python
#make sure we converted the series to the dataframe
credit4.info()
```

```python
#exported score before and after calculated mean to verify data integrity
credit4.to_csv('./data/credit4.csv')
credit2.to_csv('./data/credit2.csv')
```

```python
#Implementing the partial credit scoring based on the income bracket customers fall under.
```

```python
conditions7=[
    (application['AMT_INCOME_TOTAL']<126000),
    (application['AMT_INCOME_TOTAL']>=126000) & (application['AMT_INCOME_TOTAL']<180000),
    (application['AMT_INCOME_TOTAL']>=180000) & (application['AMT_INCOME_TOTAL']<225000),
    (application['AMT_INCOME_TOTAL']>=225000)
]
```

```python
values7=['4','6','8','10']
```

```python
application['INCOME_SCORE']=np.select(conditions7,values7)
```

```python
application['CAR_SCORE']=application['CAR_SCORE'].astype(int)
application['REALTY_SCORE']=application['REALTY_SCORE'].astype(int)
application['INCOME_TYPE_SCORE']=application['INCOME_TYPE_SCORE'].astype(int)
application['FAMILY_STATUS_SCORE']=application['FAMILY_STATUS_SCORE'].astype(int)
application['HOUSING_SCORE']=application['HOUSING_SCORE'].astype(int)
application['INCOME_SCORE']=application['INCOME_SCORE'].astype(int)
```

```python
#Begin totaling up the partial scores
```

```python
scores_list=['CAR_SCORE',
                   'REALTY_SCORE',
                   'INCOME_TYPE_SCORE',
                    'FAMILY_STATUS_SCORE',
                    'HOUSING_SCORE',
                    'INCOME_SCORE']
```

```python
application['APPLICATION_SCORE']=application[scores_list].sum(axis=1)
```

```python
application[['ID','CAR_SCORE',
                   'REALTY_SCORE',
                   'INCOME_TYPE_SCORE',
                    'FAMILY_STATUS_SCORE',
                    'HOUSING_SCORE',
                    'INCOME_SCORE', 'APPLICATION_SCORE']]
```

```python
#Now that we have calculated the credit points based on the 6 months credit payment history
#AND the credit points for the biographic data, we can merge the 2 dataframes together and calculate the overall score.
```

```python
credit_score_final=pd.merge(application,credit4,on='ID')
credit_score_final
```

```python
sum_column = credit_score_final["SCORE_SUM"] + credit_score_final["CREDIT_SCORE"]
credit_score_final['FINAL_SCORE'] = sum_column
credit_score_final
```
